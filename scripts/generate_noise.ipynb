```python
from PIL import Image
import numpy as np

def generate_rgb_noise_image(width, height):
    # Generate random RGB noise
    noise = np.random.randint(0, 256, size=(height, width, 3), dtype=np.uint8)

    # Create Pillow image from the noise
    noise_image = Image.fromarray(noise, 'RGB')

    return noise_image

def save_image(image, filename):
    image.save(filename)
    print(f"Image saved as {filename}")
```

```python

width = 1920
height =1920

# Generate RGB noise image
noise_image = generate_rgb_noise_image(width, height)

# Save the image
save_image(noise_image, "rgb_noise_image.png")
```

```python
from PIL import Image
import numpy as np

def apply_gamma_correction(image, gamma):
    # Convert image to NumPy array
    img_array = np.array(image)

    # Apply gamma correction
    corrected_array = 255 * (img_array / 255) ** gamma

    # Clip values to be in the valid range [0, 255]
    corrected_array = np.clip(corrected_array, 0, 255)

    # Convert back to uint8
    corrected_img = Image.fromarray(corrected_array.astype(np.uint8))

    return corrected_img

def save_image(image, filename):
    image.save(filename)
    print(f"Image saved as {filename}")

```

```python
# Load an image
img_name = r"C:\thtg\tg\Renderer\neural-rays\scripts\flip.ref.512spp_net.67ppd.ldr"
original_image = Image.open(img_name + ".png")

# Set the gamma value (adjust as needed)
gamma_value = .7

# Apply gamma correction
gamma_corrected_image = apply_gamma_correction(original_image, gamma_value)

# Save the corrected image
save_image(gamma_corrected_image, img_name + "_gc" + ".png")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Set random seed for reproducibility
np.random.seed(42)

# Generate random data
data = np.random.randn(1000)  # You can replace this with your own data or distribution

# Create a histogram
plt.hist(data, bins=20, density=True, alpha=0.7, color='blue', edgecolor='black')

# Fit a probability density function (PDF)
mu, std = norm.fit(data)
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, std)

# Plot the PDF
plt.plot(x, p, 'k', linewidth=2)

plt.xticks([])
plt.yticks([])
# Add labels and a title
# plt.title(f'Fit results: mu = {mu:.2f},  std = {std:.2f}')
# plt.xlabel('Value')
# plt.ylabel('Frequency')

# Save the PDF
plt.savefig('random_distribution.pdf')

# Show the plot (optional)
plt.show()
```
